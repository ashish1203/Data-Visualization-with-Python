```python
#Project of Data visualization
#1- Plot to summarize the results of a survey that was conducted to guage an audience interest in different science topics

import numpy as np
import pandas as pd

# Read csv file
df = pd.read_csv('Topic_Survey_Assignment.csv',index_col=0)
print(df.head(5)) # (1) please upload a screenshot of your dataframe with the actual numbers. 

# we are using the inline backend
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

# sort
df=df.sort_values(['Very interested'], ascending=False)
df.head()
# into percentage - decimal of 2
df= (100. * df / 2233).round(2)
print(df.head())

#plot


y_pos = ('Data Analysis / Statistics','Machine Learning','Data Visualization','Big Data (Spark / Hadoop)','Deep Learning')
bar_colors = ['#5cb85c', '#5bc0de', '#d9534f']

ax= df.plot(kind='bar',figsize=(20,8),width=0.8, color=bar_colors,edgecolor=None,linewidth=0) # bar plot, fig size, bar width, color, boundary
plt.xticks(fontsize=14) # xlabel font
plt.title("Percentage of Respondents' Interest in Data Science Areas",fontsize= 16) #title font
plt.legend(labels=df.columns,fontsize= 14) # legend font
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.axes.get_yaxis().set_visible(False)
for p in ax.patches:
    width, height = p.get_width(), p.get_height() # individual bar width and height
    #print(height)
    x, y = p.get_xy() # co-ordinates in x-y direction
    #print(x)
    ax.annotate('{:.2f}'.format(height), (x, y + height + 0.01))
#plt.show()

```
