```python
# 2- plot of choropleth map of crime rate in San Franscisco

# Based on the San Francisco crime dataset, you will find that San Francisco consists of 10 main neighborhoods, namely:

# Central,
# Southern,
# Bayview,
# Mission,
# Park,
# Richmond,
# Ingleside,
# Taraval,
# Northern, and,
# Tenderloin.

# (a) represents the total number of crimes in each neighborhood.


	# neighborhood.	count
	
# (b) As you learned in the Choropleth maps lab, you will need a GeoJSON file 

# For the map, make sure that:

# it is centred around San Francisco,
# you use a zoom level of 12,
# you use fill_color = 'YlOrRd',
# you define fill_opacity = 0.7,
# you define line_opacity=0.2, and,
# you define a legend and use the default threshold scale.


import numpy as np
import pandas as pd
# Read csv file
df = pd.read_csv('crime_rate.csv',index_col=0)
# check shape 
print(df.shape)
#drop nan element 
#df.dropna(inplace=True)
#check again shape
#print(df.shape)

#resulting object will be in descending order so that the first element is the most frequently-occurring element.
#Excludes NA values by default. 
t = df.PdDistrict.value_counts() 
print(t)
print(t.index)
# collect the values index wise under count column
t1 = pd.DataFrame(data=t.values, index=t.index, columns=['Count'])
print(t1)
#  reindex the dataframe.
t2 = t1.reindex(["CENTRAL", "NORTHERN", "PARK", "SOUTHERN", "MISSION", "TENDERLOIN", "RICHMOND", "TARAVAL", "INGLESIDE", "BAYVIEW"])
print(t2)
# When we reset the index, the old index is added as a column, and a new sequential index is used:
t3 = t2.reset_index()
print(t3)
t3.rename({'index': 'Neighborhood'}, axis='columns', inplace=True)
print(t3)



import folium
import json
with open('san-francisco.geojson') as file:
    data = json.load(file)
    
sf_zones = r'san-francisco.geojson'
#data = json.load('san-francisco.geojson').read()
#json.loads(open('config/defaults.json').read())

sf_map = folium.Map(location = [37.77, -122.42], zoom_start = 12)

#Create a layer, shaded by test scores:
sf_map.choropleth(geo_data= data,
                  fill_color='YlOrRd', 
                  fill_opacity=0.7, 
                  line_opacity=0.2,
                  threshold_scale =[0, 0.5, 0.75, 0.85, 0.9],
                  legend_name = 'Crime rate in San Franscisco')

display(sf_map)
```
